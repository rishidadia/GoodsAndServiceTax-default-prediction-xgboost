```python
df=df.drop(columns=['Transaction ID', 'Customer ID','Transaction Date','IP Address','Shipping Address','Billing Address','Customer Location'])
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['Is Fraudulent']), df['Is Fraudulent'], test_size=0.3, random_state=42, stratify=df['Is Fraudulent'])
```

PIPELINE FOR TUNING

```python
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
from xgboost import XGBClassifier
from imblearn.pipeline import Pipeline  
import category_encoders as ce
cat_cols = [
    "Payment Method",
    "Product Category",
    "Device Used"
]


pipe = Pipeline([
    # ("sampler", RandomUnderSampler(
    #     sampling_strategy=0.4,
    #     random_state=42
    # )),
    ("sampler", RandomOverSampler(
        sampling_strategy=1.0,
        random_state=42
    )),
    ("encoder", ce.TargetEncoder(cols=cat_cols)),
    ("model", XGBClassifier(
        # scale_pos_weight=(y_train==0).sum()/(y_train==1).sum(),
        objective="binary:logistic",
        eval_metric="aucpr",
        random_state=42,
        n_jobs=-1
    ))
])
```

OPTUNA BASED FINE TUNING TO OBTAIN BEST PARAMETERS

```python
from xgboost import XGBClassifier
import optuna as opt
from sklearn.metrics import confusion_matrix, average_precision_score,f1_score
from sklearn.model_selection import StratifiedKFold, cross_val_score
def objective(trial):
    param={
        'model__n_estimators':trial.suggest_int('n_estimators',100,1000),
        'model__max_depth':trial.suggest_int('max_depth',3,10),
        'model__learning_rate':trial.suggest_float('learning_rate',0.01,0.1),
        'model__min_child_weight':trial.suggest_int('min_child_weight',3,10),
        # 'model__gamma':trial.suggest_float('gamma',0,0.5),
        # 'model__subsample':trial.suggest_float('subsample',0.6,1.0),
        # 'model__scale_pos_weight':trial.suggest_int('scale_pos_weight',3,15),
        'model__objective': 'binary:logistic',
        # 'model__eval_metric': 'aucpr',
        'model__n_jobs': -1,
        'model__random_state': 42
    }
    pipe.set_params(**param)
    cv=StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
    scores = cross_val_score(
        pipe,
        X_train,        
        y_train,
        cv=cv,
        scoring="f1",
        n_jobs=-1
    )

    return scores.mean()

study = opt.create_study(direction="maximize")
study.optimize(objective, n_trials=10)

print("Best F1:", study.best_value)
print("Best params:", study.best_params)
```

FINAL PIPELINE WITH OBTAIED PARAMTERES

```python
final_pipe = Pipeline([
    ("sampler", RandomOverSampler(
        sampling_strategy=1.0,
        random_state=42
    )),
    ("encoder", ce.TargetEncoder(cols=cat_cols)),
    ("model", XGBClassifier(
        **study.best_params,
        objective="binary:logistic",
        eval_metric="aucpr",
        random_state=42,
        n_jobs=-1
    ))
])
final_pipe.fit(X_train, y_train)
# y_pred=final_pipe.predict(X_test)
```

THRESHOLD MANAGEMENT AND METRICS

```python
import numpy as np
y_prob=final_pipe.predict_proba(X_train)[:,1]
threshold=np.arange(0.0,1.0,0.05)
f1=0
best_t=0
for t in threshold:
    y_pred=(y_prob>=t).astype(int)
    if(f1_score(y_train,y_pred))>f1:
        f1=f1_score(y_train,y_pred)
        best_t=t
print(best_t, f1)
```

```python
y_pred=(final_pipe.predict_proba(X_test)[:,1]>=best_t).astype(int)
```

```python
from sklearn.metrics import confusion_matrix, average_precision_score,f1_score, accuracy_score, classification_report
print('Average Precision Score:', average_precision_score(y_test, y_pred))
print('F1 Score:', f1_score(y_test, y_pred))
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test, y_pred))
```

SHAP EXPLAINIBILTY TO DERIVE CONCLUSIONS

```python
import shap
X_test_enc = final_pipe.named_steps['encoder'].transform(X_test)
explainer = shap.Explainer(final_pipe.named_steps['model'])
X_shap=X_test_enc.sample(2000, random_state=42)
shap_values=explainer.shap_values(X_shap)
shap.summary_plot(
    shap_values,
    X_shap,
    max_display=20
)
```
